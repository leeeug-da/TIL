### **ğŸ”¸ì„œìš¸ì‹œì˜ ì•½êµ­ ì¤‘ 'ì›”ìš”ì¼ ì•¼ê°„'ì— ìš´ì˜í•˜ëŠ” ì•½êµ­ ì •ë³´ë¥¼ ì¶”ì¶œí•´ë³´ì**
- 'ì•¼ê°„'ì˜ ê¸°ì¤€ : ì˜¤í›„ 9ì‹œ ì´í›„

```python
# ì²«ë²ˆì§¸ ì•½êµ­ì— ëŒ€í•´ ì‹œê°„ì •ë³´ í™•ì¸í•´ë³´ê¸° 
# res_fin[0] => dutytime1c, dutytime1s, dutytime2c, dutytime2s ..
# dutytime*c = close
# dutytime*s = start 
# * = 1,2,3,4,5,6 = ì›”,í™”,ìˆ˜,ëª©,ê¸ˆ,í†  ë¡œ ì¶”ì •

# ì›”ìš”ì¼ ì˜ì—… ì¢…ë£Œì‹œê°„ì´ 21:00ë¶„ ì´í›„ì¸ì§€ í™•ì¸
# ì›”ìš”ì¼ ì¢…ë£Œ ì‹œê°„ íƒœê·¸ = <dutytime1c>
int(res_fin[0].dutytime1c.text) > 2100
```

```python
name = []
adr = []
tel = []
cl_time = []
count = 0

for item in res_fin :
    tmp_time = item.find('dutytime1c')
    # ì›”ìš”ì¼ íœ´ë¬´ì¸ ì•½êµ­ì€ íƒœê·¸ê°€ ì—†ì„ ìˆ˜ ìˆìŒ
    if tmp_time != None :
        c_time = int(tmp_time.text)
        if c_time > 2100 :
            count = count + 1
            name.append(item.find('dutyname').text)
            adr.append(item.find('dutyaddr').text)
            tel.append(item.find('dutytel1').text)
            cl_time.append(c_time)
```

```python
print('ì›”ìš”ì¼ ì•¼ê°„ì˜ì—… ì•½êµ­ ê°¯ìˆ˜: ', count)
print('ëª¨ë“  ë°ì´í„°ê°€ ë§ê²Œ ë“¤ì–´ì™”ëŠ”ì§€ í™•ì¸: ', len(name), len(adr), len(tel), len(cl_time))
```

```python
df_mon_night = pd.DataFrame({'ì•½êµ­ëª…':name, 'ì£¼ì†Œ':adr, 'ì „í™”ë²ˆí˜¸':tel, 'ì˜ì—…ì¢…ë£Œì‹œê°„':cl_time})
df_mon_night
```

```python
# íŒŒë¼ë¯¸í„° ìƒì„±
paramset = '?servicekey='+servicekey + \
            '&Q0='+Q0 + \
            '&ORD='+ORD + '&numOfRows='+numOfRows
paramset
```

```python
# URL ìš”ì²­ ë° bs4 ê°ì²´ ìƒì„±
result = requests.get(url)
bs_obj = BeautifulSoup(result.content, 'html.parser')
```

```python
# 'item' ìœ¼ë¡œ ë¶„ë¦¬ë˜ì–´ ìˆëŠ” ê²ƒì„ í™•ì¸ 
res_fin = bs_obj.find_all('item')
print('ì„œìš¸ì‹œì˜ ëª¨ë“  ì•½êµ­ ê°¯ìˆ˜: ', len(res_fin))
```

```python
# res_fin[0]ì„ ì˜ˆì‹œë¡œ ì½”ë“œ ë§Œë“¤ì–´ë³´ê¸° 
# dutytime*c = close
# dutytime*s = start 
# * = 1,2,3,4,5,6 = ì›”,í™”,ìˆ˜,ëª©,ê¸ˆ,í†  ë¡œ ì¶”ì •

# ì•½êµ­ëª… í™•ì¸ ì½”ë“œ
res_fin[0].find('dutyname').text
# ì£¼ì†Œ í™•ì¸ ì½”ë“œ
res_fin[0].find('dutyaddr').text
# ì „í™”ë²ˆí˜¸ í™•ì¸ ì½”ë“œ
res_fin[0].find('dutytel1').text

# í† ìš”ì¼ ì˜ì—… í•˜ëŠ”ì§€ ì—¬ë¶€ í™•ì¸ ì½”ë“œ # dutytime6
if (res_fin[0].find('dutytime6s') == None) == False:
    print('Y')
    # í† ìš”ì¼ ì¢…ë£Œì‹œê°„ í™•ì¸ ì½”ë“œ
    print(res_fin[0].find('dutytime6c').text)
else:
    print('N')

# ì¼ìš”ì¼ ì˜ì—… í•˜ëŠ”ì§€ ì—¬ë¶€ í™•ì¸ ì½”ë“œ
if (res_fin[0].find('dutytime7s') == None) == False:
    print('Y')
    # ì¼ìš”ì¼ ì¢…ë£Œì‹œê°„ í™•ì¸ ì½”ë“œ
    print(res_fin[0].find('dutytime7c').text)
else:
    print('N')
```

```python
# ì•½êµ­ëª…, ì£¼ì†Œ, ì „í™”ë²ˆí˜¸, í† ìš”ì¼ì˜ì—…ì—¬ë¶€, ì¼ìš”ì¼ì˜ì—…ì—¬ë¶€, í† ìš”ì¼ì¢…ë£Œì‹œê°„, ì¼ìš”ì¼ì¢…ë£Œì‹œê°„
# ëª¨ë“  dataì— ì ìš© 
name = []
adr = []
tel = []
open_sat = []
ctime_sat = []
open_sun = []
ctime_sun = []
count = 0

for item in res_fin :
    name.append(item.find('dutyname').text)
    adr.append(item.find('dutyaddr').text)
    tel.append(item.find('dutytel1').text)

    # í† ìš”ì¼ ì˜ì—… í•˜ëŠ”ì§€ ì—¬ë¶€ í™•ì¸ ì½”ë“œ # dutytime6
    if (item.find('dutytime6s') == None) == False:
        open_sat.append('Y')
        # í† ìš”ì¼ ì¢…ë£Œì‹œê°„ í™•ì¸ ì½”ë“œ
        ctime_sat.append(item.find('dutytime6c').text)
    else:
        open_sat.append('N')
        ctime_sat.append('N')

    # ì¼ìš”ì¼ ì˜ì—… í•˜ëŠ”ì§€ ì—¬ë¶€ í™•ì¸ ì½”ë“œ
    if (item.find('dutytime7s') == None) == False:
        open_sun.append('Y')
        # ì¼ìš”ì¼ ì¢…ë£Œì‹œê°„ í™•ì¸ ì½”ë“œ
        ctime_sun.append(item.find('dutytime7c').text)
    else:
        open_sun.append('N')
        ctime_sun.append('N')
        
    count = count + 1
```

```python
print('ì„œìš¸ì‹œ ì•½êµ­ ê°¯ìˆ˜: ', count)
print('ëª¨ë“  ë°ì´í„°ê°€ ë§ê²Œ ë“¤ì–´ì™”ëŠ”ì§€ í™•ì¸: ', len(name), len(adr), len(tel))
print('í† ìš”ì¼ ì˜ì—… ì•½êµ­ ê°¯ìˆ˜ê°€ ì„œë¡œ ë§ëŠ”ì§€ í™•ì¸: ', open_sat.count('N'), ctime_sat.count('N'))
print('ì¼ìš”ì¼ ì˜ì—… ì•½êµ­ ê°¯ìˆ˜ê°€ ì„œë¡œ ë§ëŠ”ì§€ í™•ì¸: ', open_sun.count('N'), ctime_sun.count('N'))
```

```python
# dfë¡œ ì €ì¥ 
df_satsun = pd.DataFrame({'ì•½êµ­ëª…':name, 'ì£¼ì†Œ':adr, 'ì „í™”ë²ˆí˜¸':tel, 'í† ìš”ì¼ì˜ì—…ì—¬ë¶€':open_sat, 
                          'í† ìš”ì¼ì¢…ë£Œì‹œê°„':ctime_sat, 'ì¼ìš”ì¼ì˜ì—…ì—¬ë¶€':open_sun, 'ì¼ìš”ì¼ì¢…ë£Œì‹œê°„':ctime_sun})
df_satsun.tail()
```

```python
# ì €ì¥
df_satsun.to_csv('./crawl_data/ì„œìš¸ì‹œ_ì£¼ë§ì˜ì—…ì•½êµ­.csv')
```
