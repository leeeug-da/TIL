### Packages

```python
import pandas as pd
```

### 축제 데이터

```python
fest_g = pd.read_excel('./data_info/축제_경기도_데이터.xls')
fest_s = pd.read_excel('./data_info/축제_서울_데이터.xls')
fest_i = pd.read_excel('./data_info/축제_인천_데이터.xls')
```

```python
# 각 데이터 columns 확인
print(fest_g.columns)
print(fest_s.columns)
print(fest_i.columns)
```

```python
# 각 데이터 개수 확인
print(fest_g['명칭'].count())
print(fest_s['명칭'].count())
print(fest_i['명칭'].count())
```

```python
fest_df = pd.concat([fest_g, fest_s, fest_i])
fest_df.info()
```

```python
fest_df.sort_values(by='명칭', inplace=True)
fest_df.head()
```

```python
fest_df.to_excel('./축제_데이터.xlsx')
```
