# 키워드별 상위 20개의 이미지 경로 확인

```python
import pandas as pd
```

```python
keyrat = pd.read_excel('Top20_keyrat.xlsx')
```

```python
keyrat.head()
```

```python
imgpath = pd.read_excel('tour_imgpath.xlsx')
```

```python
imgpath.head()
```

```python
keyrat.info()
imgpath.info()
```

```python
merged_df = pd.merge(keyrat, imgpath, on='tour_id', how='inner')
merged_df.head()
```

```python
merged_df.info()
```

```python
unique_rows = merged_df.drop_duplicates(subset='tour_id')
print(unique_rows[['tour_id', 'img_path']])
```
