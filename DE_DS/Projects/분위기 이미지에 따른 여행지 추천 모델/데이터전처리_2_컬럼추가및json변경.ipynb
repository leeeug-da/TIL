### 🔴 축제

```python
fest_g = pd.read_excel('./data_info/축제공연행사/축제공연행사_축제_경기도.xls')
fest_s = pd.read_excel('./data_info/축제공연행사/축제공연행사_축제_서울.xls')
fest_i = pd.read_excel('./data_info/축제공연행사/축제공연행사_축제_인천.xls')

# '지역' 컬럼 추가 
fest_g['지역'] = '경기도'
fest_s['지역'] = '서울'
fest_i['지역'] = '인천'

# '분류' 컬럼 추가
fest_g['분류'] = '축제'
fest_s['분류'] = '축제'
fest_i['분류'] = '축제'

# 각 데이터 columns 확인
print('축제_경기도 컬럼: ', fest_g.columns)
print('축제_서울 컬럼: ', fest_s.columns)
print('축제_인천 컬럼: ', fest_i.columns)

# 각 데이터 개수 확인
print('축제_경기도 데이터 갯수: ', fest_g['명칭'].count())
print('축제_서울 데이터 갯수: ', fest_s['명칭'].count())
print('축제_인천 데이터 갯수: ', fest_i['명칭'].count())
```

```python
# 데이터 합치기
fest_df = pd.concat([fest_g, fest_s, fest_i])
# 오름차순 정렬
fest_df.sort_values(by='명칭', inplace=True)
#확인
display(fest_df.head())
display(fest_df.info())
```

```python
# 내보내기
fest_df.to_json('./data_info/중분류데이터/축제공연행사_축제.json', orient='records', lines=True)
fest_df.to_csv('./data_info/중분류데이터/축제공연행사_축제.csv', index=False)
```
