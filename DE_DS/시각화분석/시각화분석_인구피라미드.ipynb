```python
# Îç∞Ïù¥ÌÑ∞ Î∂àÎü¨Ïò§Í∏∞
df_m = pd.read_excel('./data/201310_201310_Ïó∞Î†πÎ≥ÑÏù∏Íµ¨ÌòÑÌô©_ÏõîÍ∞Ñ.xlsx', skiprows=3, 
              index_col='ÌñâÏ†ïÍ∏∞Í¥Ä', usecols='B, E:Y')
df_m.head(3)
```

```python
# Îç∞Ïù¥ÌÑ∞ Î∂àÎü¨Ïò§Í∏∞
df_w = pd.read_excel('./data/201310_201310_Ïó∞Î†πÎ≥ÑÏù∏Íµ¨ÌòÑÌô©_ÏõîÍ∞Ñ.xlsx', skiprows=3,
                    index_col='ÌñâÏ†ïÍ∏∞Í¥Ä', usecols='B, AB:AV')
df_w.head(3)
```
