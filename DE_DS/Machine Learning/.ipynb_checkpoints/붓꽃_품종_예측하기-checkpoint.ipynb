## 사이킷런 모듈 임포트

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

## 데이터 세트 로딩

```python
iris = load_iris()
iris
```

## Feature/Target 구성 보기위해 DF로 변환

[Original Data]
- iris : original data


[Feature]
- 'data' : feature data
- 'feature_names' : feature name


[Label]
- 'target' : target data
- 'target_names' : target name

```python
import pandas as pd
```

```python
# Feature Data
iris_data = iris.data
iris_data
```

```python
# Feature Names
iris.feature_names
```

```python
# Target Data
iris_label = iris.target
iris_label
```

```python
# Target Name
iris.target_names
```

```python
iris_df = pd.DataFrame(data=iris_data, columns=iris.feature_names)
```

```python
iris_df['label'] = iris.target
```
