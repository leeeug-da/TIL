```python
N = int(input())
paper = [[0 for i in range(100)] for j in range(100)]

for _ in range(N):
    x, y = map(int, input().split())
    for xpoint in range(x, x+10):
        for ypoint in range(y, y+10):
            paper[xpoint][ypoint] = 1

counts = 0

for points in paper:
    counts += points.count(1)

print(counts)
```
