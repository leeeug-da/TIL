```python
# 2    ,       
array = [[1]]  
current_num = 2 

while current_num <= 100:  
    for i in range(1,10):
        array.append(list(range(current_num, current_num + 6*i))) 
        current_num += 6*i 

N = int(input())
for i in range(len(array)):
    if N in array[i]:
        print(i+1)
        break
```
