### âœï¸ ì²˜ìŒì— í‘¼ ë°©ë²•

```python
A, B, C = map(int, input().split())

origin = [A, B, C]
unique = []
repeat = []

for org in origin:
    if org not in unique:
        unique.append(org)
    else:
        repeat.append(org)

# ëª¨ë‘ ë‹¤ë¥¸ ëˆˆì´ ë‚˜ì˜¤ëŠ” ê²½ìš° - repeat[0ê°œ]
largest = origin[0]

if len(repeat) == 0:
    for org in origin:
        if org > largest:
            largest = org
    print(largest*100)

# ê°™ì€ ëˆˆì´ 3ê°œê°€ ë‚˜ì˜¤ëŠ” ê²½ìš° - repeat[2ê°œ]
elif len(repeat) == 2:
    unique = unique[0]
    print(10000+unique*1000)

# ê°™ì€ ëˆˆì´ 2ê°œë§Œ ë‚˜ì˜¤ëŠ” ê²½ìš° - repeat[1ê°œ]
else:
    repeat = repeat[0]
    print(1000+repeat*100)
```
